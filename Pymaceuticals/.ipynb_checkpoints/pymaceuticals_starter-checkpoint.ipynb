## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined = pd.merge(mouse_metadata, study_results, on='Mouse ID', how='outer')
# Display the data table for preview
combined
```

```python
# Checking the number of mice.
total_mice_number = combined['Mouse ID'].nunique()
total_mice_number
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated = combined[combined.duplicated(['Mouse ID','Timepoint'])]
duplicated
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
combined.loc[combined['Mouse ID']=="g989",:]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
combined_clean = combined.loc[combined['Mouse ID']!="g989",:]
combined_clean
```

```python
# Checking the number of mice in the clean DataFrame.
combined_clean['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
drug_groupby = combined_clean.groupby('Drug Regimen')
tumor_vol_mean = drug_groupby['Tumor Volume (mm3)'].mean()
tumor_vol_median = drug_groupby['Tumor Volume (mm3)'].median()
tumor_vol_var = drug_groupby['Tumor Volume (mm3)'].var()
tumor_vol_std = drug_groupby['Tumor Volume (mm3)'].std()
tumor_vol_sem = drug_groupby['Tumor Volume (mm3)'].sem()

summary_df = pd.DataFrame({"Tumor Volume (mm3) Mean":tumor_vol_mean,
                           "Tumor Volume (mm3) Median":tumor_vol_median,
                           "Tumor Volume (mm3) Variance":tumor_vol_var,
                           "Tumor Volume (mm3) Standard Deviation":tumor_vol_std,
                           "Tumor Volume (mm3) Standard Error":tumor_vol_sem
})
summary_df.style.set_table_styles([dict(selector="th",props=[('max-width', '80px')])])
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

summary_df_2 = drug_groupby['Tumor Volume (mm3)'].agg(['mean','median','var','std','sem'])
summary_df_2
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
Capomulin_df = combined_clean.loc[(combined_clean['Drug Regimen']=="Capomulin"),['Mouse ID','Timepoint','Tumor Volume (mm3)']]

Capomulin_mice = Capomulin_df.groupby(['Mouse ID','Timepoint'])
Capomulin_mice_sum = Capomulin_mice.sum()

mouseid = 'b128'
one_mouse_data = Capomulin_mice_sum.loc[mouseid]
plot = one_mouse_data.plot(title="Tumor Volume vs. Time of mouse " +mouseid)
plot.set_xlabel("Time")
plot.set_ylabel("Tumor Volume (mm3)")
plot.set_xlim(0,45)
plot.set_ylim(min(one_mouse_data['Tumor Volume (mm3)'])-1, max(one_mouse_data['Tumor Volume (mm3)'])+1)
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

Capomulin_df_2 = combined_clean.loc[(combined_clean['Drug Regimen']=="Capomulin"),['Mouse ID','Weight (g)','Tumor Volume (mm3)']]
Capomulin_df_2

Capomulin_weight = Capomulin_df_2.groupby('Mouse ID')
Capomulin_weight_mean = Capomulin_weight.mean()

plot2=Capomulin_weight_mean.plot(kind='scatter', x='Weight (g)', y='Tumor Volume (mm3)', title="Average Tumor Volume (mm3) vs. Mouse Weight (g)")

plot2.set_xlabel("Mouse Weight (g)")
plot2.set_ylabel("Average Tumor Volume (mm3)")
plot2.set_xlim(min(Capomulin_weight_mean['Weight (g)'])-1, max(Capomulin_weight_mean['Weight (g)'])+1)
plot2.set_ylim(min(Capomulin_weight_mean['Tumor Volume (mm3)'])-1, max(Capomulin_weight_mean['Tumor Volume (mm3)'])+1)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation = st.pearsonr(Capomulin_weight_mean['Weight (g)'],Capomulin_weight_mean['Tumor Volume (mm3)'])
print(f"The correlation between mouse weight and average tumor volume is {round(correlation[0],2)}")
```

```python
x_values = Capomulin_weight_mean['Weight (g)']
y_values = Capomulin_weight_mean['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,37), fontsize=15,color="red")
plt.title("Average Tumor Volume (mm3) vs. Mouse Weight (g)")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
print(f"The r-squared is: {rvalue**2}")
```
